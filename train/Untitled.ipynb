```python
import numpy as np
import torch as th

```

```python
train_set = th.rand(100, 5)
print(train_set.shape)

test_set = th.sum(train_set, axis=1)
test_set = th.unsqueeze(test_set, 1)
print(test_set.shape)

# train_set = th.form_numpy(train_set)
# test_set = th.form_numpy(test_set)
```

```python

import torch.nn as nn

class SumNet(nn.Module):
    def __init__(self):
        super(SumNet, self).__init__()
        self.layer_1 = nn.Linear(5, 5)
        self.relu = nn.ReLU()
        self.layer_2 = nn.Linear(5, 5) 
        self.layer_3 = nn.Linear(5, 5) 
        self.layer_4 = nn.Linear(5, 1)
        
    def forward(self, input_vector):
        output = self.layer_1(input_vector)
        output = self.relu(output)
        output = self.layer_2(input_vector)
        output = self.relu(output)
        output = self.layer_3(input_vector)
        output = self.relu(output)
        output = self.layer_4(input_vector)
        return output


net = SumNet()
optim = th.optim.Adam(net.parameters(), lr=1e-4, weight_decay=1e-5)

for j in range(100):
    for i, input_vector in enumerate(train_set, 0):
        pred = net.forward(input_vector)

        loss = nn.functional.l1_loss(test_set[i], pred)


        optim.zero_grad()
        loss.backward() #th.ones_like(loss))
        optim.step()

        print(loss)
    

# for i in range(1):

#     current_loss = []
#     self.epochs += 1

#     random_img = random_target = random_pred = None

#     for j, data in enumerate(trainloader, 0):
#         img, target = data

#         target = target.to(self.device)
#         img = img.to(self.device)
#         x_predicted = self.model.forward(img)

#         if j == 0:
#             random_img = img.cpu().detach().numpy()
#             random_target = target.cpu().detach().numpy()
#             random_pred = x_predicted.cpu().detach().numpy()

#         loss = self.criterion(x_predicted, target)

#         self.optimizer.zero_grad()
#         loss.backward() #th.ones_like(loss))
#         self.optimizer.step()
    
```
